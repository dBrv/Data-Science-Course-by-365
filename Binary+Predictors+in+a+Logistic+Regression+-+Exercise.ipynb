# Binary Predictors in a Logistic Regression

Using the same code as in the previous exercise, find the odds of 'duration'. 

What do they tell you?

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from scipy import stats
stats.chisqprob = lambda chisq, df: stats.chi2.sf(chisq, df)
```

## Load the data

Load the ‘Bank_data.csv’ dataset.

```python
raw_data = pd.read_csv("Bank_data.csv")
raw_data
```

```python
data = raw_data.copy()
data = data.drop(["Unnamed: 0"], axis = 1)
data["y"] = data["y"].map({'yes': 1, "no": 0})
data
```

```python
data.describe()
```

### Declare the dependent and independent variables

Use 'duration' as the independet variable.

```python
y = data["y"]
x1 = data['duration']
```

### Simple Logistic Regression

Run the regression.

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
results_log.summary()
```

```python
plt.scatter(x1,y,color= 'C0')
plt.xlabel('duration', fontsize = 20)
plt.ylabel('subscription', fontsize = 20)
plt.show()
```

### Find the odds of duration

```python
np.exp(0.0051)
```

Even though the duration is a signifficant predictor, a change in just 1 day would barely affect the regression. The coefficient itself tells us that.
Note that the data is not standardized and duration is a feature of a relatively big order of magnitude.
